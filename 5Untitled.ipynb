```python
# Sales Data Analysis

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
# Error: The file 'sales_data.csv' doesn't exist in the current directory
# Option 1: Provide the correct path to your CSV file
# df = pd.read_csv("correct_path/sales_data.csv")

# Option 2: Create sample data for demonstration
# Creating sample data instead of loading from a non-existent file
data = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product A', 'Product B', 
                'Product C', 'Product A', 'Product B', 'Product C', 'Product A'],
    'Region': ['North', 'South', 'East', 'West', 'North', 
               'South', 'East', 'West', 'North', 'South'],
    'Revenue': [1200, 1500, 1300, 1400, 1600, 
                1100, 1700, 1900, 1250, 1350]
}
df = pd.DataFrame(data)

# Basic info
print(df.head())
print(df.describe())

# Total revenue overall
total_revenue = df["Revenue"].sum()
print("Total Revenue:", total_revenue)

# Revenue by product
revenue_product = df.groupby("Product")["Revenue"].sum().sort_values(ascending=False)
print(revenue_product)

# Revenue by region
revenue_region = df.groupby("Region")["Revenue"].sum()
print(revenue_region)

# Visualization: Revenue by product
plt.figure(figsize=(8,5))
revenue_product.plot(kind="bar", color="skyblue")
plt.title("Revenue by Product")
plt.ylabel("Total Revenue (RM)")
plt.show()

# Visualization: Revenue by region
plt.figure(figsize=(6,4))
revenue_region.plot(kind="pie", autopct="%1.1f%%")
plt.title("Revenue Distribution by Region")
plt.ylabel("")
plt.show()
```
