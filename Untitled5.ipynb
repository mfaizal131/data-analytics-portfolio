```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Dataset
data = {
    "Income Range (RM)": [
        "0-1999", "2000-2999", "3000-3999", "4000-4999", "5000-5999",
        "6000-6999", "7000-7999", "8000-8999", "9000-9999", "10000-10999",
        "11000-11999", "12000-12999", "13000-13999", "14000-14999", ">=15000"
    ],
    "Neighbourhood A (%)": [3, 8, 17, 14, 12, 9, 7, 6, 5, 4, 3, 2, 2, 2, 5],
    "Neighbourhood B (%)": [9, 12, 16, 14, 10, 8, 6, 5, 3, 3, 2, 1, 1, 1, 6],
    "Neighbourhood C (%)": [2, 4, 9, 12, 11, 10, 8, 7, 6, 5, 4, 3, 3, 2, 13],
}

df = pd.DataFrame(data)
df.head()
```

```python
midpoints = [
    1000, 2500, 3500, 4500, 5500,
    6500, 7500, 8500, 9500, 10500,
    11500, 12500, 13500, 14500, 16000
]

def weighted_median(data, weights):
    data, weights = np.array(data), np.array(weights)
    sorter = np.argsort(data)
    data, weights = data[sorter], weights[sorter]
    cumsum = np.cumsum(weights)
    cutoff = weights.sum() / 2.0
    return data[np.searchsorted(cumsum, cutoff)]

median_A = weighted_median(midpoints, df["Neighbourhood A (%)"])
median_B = weighted_median(midpoints, df["Neighbourhood B (%)"])
median_C = weighted_median(midpoints, df["Neighbourhood C (%)"])

print(f"Median A: RM{median_A:,.0f}")
print(f"Median B: RM{median_B:,.0f}")
print(f"Median C: RM{median_C:,.0f}")
```

```python
df.set_index("Income Range (RM)")[["Neighbourhood A (%)", "Neighbourhood B (%)", "Neighbourhood C (%)"]].plot(
    kind="bar", figsize=(12,6))
plt.title("Income Distribution by Neighbourhood")
plt.ylabel("Percentage of Households (%)")
plt.xlabel("Income Range (RM)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(["A", "B", "C"], [median_A, median_B, median_C], color=["skyblue","salmon","lightgreen"])
plt.title("Estimated Median Incomes")
plt.ylabel("RM")
plt.show()
```
